# **Polynomial Logistic Regression**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = np.random.randint(-100, 10, 100)
y = np.random.randint(-100, 10, 100)
```

```python
df = pd.DataFrame({'x': x, 'y': y})
```

```python
plt.scatter(df['x'], df['y'], c= df['x'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Points')
plt.show()
```

```python
poly = PolynomialFeatures(degree=3)
transformed = poly.fit_transform(df['x'].values.reshape(-1,1),df['y'].values.reshape(-1,1))
```
